```python
import requests
import json
import pandas as pd
```

```python
base_url = 'http://127.0.0.1:8000'

endpoint_raw = '/get_raw_data'
endpoint_sentiment_table = '/get_sentiment_data'
endpoint_sentiment = '/vader_sentiment'

hashtag = 'legostarwars'

num_tweets_to_analyse = 100
```

```python
params_raw = {'hashtag': hashtag, 'num_tweets': num_tweets_to_analyse}
response_raw = requests.get(base_url + endpoint_raw, params=params_raw)
```

```python
response_raw_json = response_raw.json()
df_response_raw = pd.DataFrame.from_dict(response_raw_json)
df_response_raw.to_clipboard
json_dataframe_raw = df_response_raw.to_json(orient='records')
print(df_response_raw)
```

```python
params_sentiment = {'data': json_dataframe_raw}
response_sentiment = requests.get(base_url + endpoint_sentiment, params=params_sentiment)
print(response_sentiment.json())
```

```python
params_sentiment_table = {'data': json_dataframe_raw}
response_sentiment_table_dict = requests.get(base_url + endpoint_sentiment_table, params=params_sentiment_table)
response_sentiment_table_json = response_sentiment_table_dict.json()
print(response_sentiment_table_json)
df_sentiment_table = pd.DataFrame.from_dict(response_sentiment_table_json)
df_sentiment_table.to_clipboard
print(df_sentiment_table)
```
